```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('2015.csv')
df.head(10)
```

##### **Menentukan Tujuan, User dan Tone**
##### **Tujuan** : Menyampaikan informasi penting dan menganalisis hubungan antar faktor-faktor.
##### **User** : Pembuat kebijakan dan pemerintah, akademisi dan peneliti.
##### **Tone** : _informative & Neutral_ atau memberikan informasi berbasis data tanpa menyudutkan atau mendukung wilayah dan negara tertentu.

```python
# 1. Visualisasi Distribusi Skor Kebahagiaan Berdasarkan Region
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y='Happiness Score', data=df, palette='Set2')
plt.xticks(rotation=90)
plt.title('Boxplot Distribusi Skor Kebahagiaan Berdasarkan Wilayah', fontsize=14)
plt.xlabel('Wilayah', fontsize=12)
plt.ylabel('Skor Kebahagiaan', fontsize=12)
plt.show()
```

Dari boxplot diatas, dapat disimpulkan bahwa wilayah dengan kebahagian tertinggi adalah _Western Europe_ dan _North America_ yang mana secara konsisten menunjukkan tingkat kebahagiaan yang tinggi dengan distribusi yang sempit. Dan _Sub-Saharan Africa_ serta _Southern Asia_ merupakan wilayah dengan kebahagiaan rendah dan distribusi yang lebih lebar.

```python
# 2. Visualisasi Hubungan GDP per Kapita dengan Skor Kebahagiaan
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Economy (GDP per Capita)', y='Happiness Score', data=df, hue='Region', palette='Set2', s=100)
plt.title('Hubungan Economy (GDP per Kapita) dengan Skor Kebahagiaan', fontsize=14)
plt.xlabel('Ekonomi (GDP per Kapita)', fontsize=12)
plt.ylabel('Skor Kebahagiaan', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1))
plt.show()
```

Dari plot ini, dapat disimpulkan bahwa Ekonomi berkontribusi secara signifikan terhadap skor kebahagiaan seseorang. Namun, ekonomi bukan satu-satunya yang menentukan kebahagiaan.

```python
# 3. Visualisasi Korelasi antar Indikator Kebahagiaan
plt.figure(figsize=(10, 8))
correlation = df[['Happiness Score', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Korelasi antar Indikator Kebahagiaan', fontsize=14)
plt.show()
```

Dari matriks korelasi ini, dapat dilihat bahwa nilai yang semakin mendekati angka 1 maka dapat dinyatakan bahwa faktor tersebut saling berhubungan. Cntohnya faktor kesehatan terhadap skor kebahagiaan yang memiliki nilai 0.72 yang mendekati 1.
